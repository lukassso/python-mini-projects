## Rysowanie wykresów

1. Narysuj wykres liniowy funkcji $f(x) = x^2$ dla $x$ należącego do przedziału [-10, 10].

---
(czas: 6 min.)

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python


x = np.linspace(-10, 10, 400)
y = x**2

plt.plot(x,y)

plt.title('Wykres funkcji $f(x)= x^2$')
plt.xlabel('$x$')
plt.ylabel('$y$')

plt.show()
```

2. Narysuj na jednym wykresie wykresy liniowe funkcji $f(x) = \text{sin}(x)$ oraz $f(x) = \text{cos}(x)$

(czas: 6 min.)

```python
x = np.linspace(-2 * np.pi, 2 * np.pi, 400)

y_sin = np.sin(x)
y_cos = np.cos(x)

plt.figure(figsize=(10, 6)) 
plt.plot(x, y_sin, label='sin(x)', color='blue') 
plt.plot(x, y_cos, label='cos(x)', color='red') 

plt.title('Wykresy funkcji $f(x) = \\sin(x)$ oraz $f(x) = \\cos(x)$')
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5) 
plt.axvline(0, color='black', linewidth=0.5) 
plt.legend()  

plt.show()


```

3. Oba powyższe wykresy zmodyfikuj tak aby były wykresami punktowymi.

---
(czas: 2 min.)

```python
x = np.linspace(-10, 10, 200)
y = x**2

plt.plot(x, y, 'o', markersize=2) # 'o' oznacza kropkę jako marker, markersize=2 dostosowuje jej rozmiar

plt.title('Wykres punktowy funkcji $f(x)= x^2$')
plt.xlabel('$x$')
plt.ylabel('$y$')

plt.show()

x = np.linspace(-2 * np.pi, 2 * np.pi, 200)

y_sin = np.sin(x)
y_cos = np.cos(x)

plt.figure(figsize=(10, 6)) 
plt.plot(x, y_sin, 'o', label='sin(x)') 
plt.plot(x, y_cos, 'o', label='cos(x)') 

plt.title('Wykresy punktowe funkcji $f(x) = \\sin(x)$ oraz $f(x) = \\cos(x)$')
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5) 
plt.axvline(0, color='black', linewidth=0.5) 
plt.legend()  

plt.show()
```

9. Sprawdź ile jest wystąpień samochodów spalających poszczególne rodzaje paliwa. Narysuj wykres słupkowy, który to zobrazuje.

---
(czas: 7 min.)

```python
fuel_counts = df['fuel'].value_counts()

print("Counts of fuel types:")
print(fuel_counts)



plt.figure(figsize=(7, 3))
plt.bar(fuel_counts.index, fuel_counts.values, color=['skyblue', 'lightcoral'])
plt.title('Liczba samochodów według rodzaju paliwa')
plt.xlabel('Rodzaj paliwa')
plt.ylabel('Liczba samochodów')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--')
plt.tight_layout()
plt.savefig('fuel_counts_bar_chart.png')
```

10. Narysuj dwa wykresy słupkowe. Przedstaw na nich:
- średnią cenę
- liczebność

z podziałem na samochody z automatyczną oraz manualną skrzynią biegów

---
(czas: 5 min.)

```python

# Convert relevant columns to numeric type
df['price'] = pd.to_numeric(df['price'], errors='coerce')
df['gearbox_is_manual'] = df['gearbox_is_manual'].replace({True: 'Manualna', False: 'Automatyczna'})

# Group by gearbox type and calculate average price and count
avg_price = df.groupby('gearbox_is_manual')['price'].mean().sort_index()
counts = df['gearbox_is_manual'].value_counts().sort_index()

# Create a figure with two subplots
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(16, 6))

# Plot the average price bar chart
avg_price.plot(kind='bar', ax=axes[0], color=['darkorange', 'skyblue'])
axes[0].set_title('Średnia cena samochodu\nwedług skrzyni biegów', fontsize=16)
axes[0].set_xlabel('Skrzynia biegów')
axes[0].set_ylabel('Średnia cena (PLN)')
axes[0].tick_params(axis='x', rotation=0)
for i, v in enumerate(avg_price):
    axes[0].text(i, v + 100, f'{v:.2f}', ha='center')

# Plot the counts bar chart
counts.plot(kind='bar', ax=axes[1], color=['darkorange', 'skyblue'])
axes[1].set_title('Liczebność samochodów\nwedług skrzyni biegów', fontsize=16)
axes[1].set_xlabel('Skrzynia biegów')
axes[1].set_ylabel('Liczba samochodów')
axes[1].tick_params(axis='x', rotation=0)
for i, v in enumerate(counts):
    axes[1].text(i, v + 0.1, str(v), ha='center')

plt.tight_layout()
plt.savefig('gearbox_stats.png')
```
