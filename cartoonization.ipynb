```python
from ml4a import image
from ml4a.models import cartoonization

img = image.load_image('https://www.irishtimes.com/polopoly_fs/1.3170107.1501253408!/image/image.jpg_gen/derivatives/ratio_1x1_w1200/image.jpg')
img_toon = cartoonization.run(img)
image.display([img, img_toon])
```
